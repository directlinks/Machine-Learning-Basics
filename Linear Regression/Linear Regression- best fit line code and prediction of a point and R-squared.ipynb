# Math behind Linear Regression

y=mx+b

## calculate the best fit line i.e "m" and calculate the intercept "b".

### best fit line tells whether the 2 variables are co-related any how or not

```python
########################################### calculating the best fit slope #####################################

x=np.array([1,2,3,4,5,6])
y=np.array([5,4,6,5,6,7])

def best_fit_line(x,y):
    
    m= ( ((np.mean(x)*np.mean(y))-(np.mean(x*y))) /
        (np.mean(x)**2- np.mean(x**2)) )
    
    return m

slope=best_fit_line(x,y)
print slope

def intercept(slope,y,x):
    b= (mean(y) - (slope*mean(x)))
    
    return b

b=intercept(slope,y,x)

print b
    
```

```python
regression_line=[(slope*i)+b for i in x]

plt.scatter(x,y)
plt.plot(x,regression_line)
plt.show()
```

### we have succesfully created a Linear Regression model using the best fit line and intercept

#### let us predict a point on the same graph


```python
predict_x=8
predict_y= slope*predict_x+b

plt.scatter(x,y)
plt.scatter(predict_x,predict_y, color="green")
plt.plot(x,regression_line)

plt.show()
```

#### Green point is the predicted point based on the previous points

# Calculation of R Squared Error:-

#### R-squared is a statistical measure of how close the data are to the fitted regression line

### formula => square(R)= 1- (Square_error_regression_line / Square_Error__mean_y) 


```python
def square_error(regression_line, y):
    
    return sum((regression_line-y)**2)

def coefficient(regression_line,y):
    mean_list=[mean(y) for i in y]   # mean of y---> type list
    square_error_regression_line=square_error(regression_line,y)
    square_error_mean_y=square_error(y,mean_list)
    return 1-(square_error_regression_line/square_error_mean_y)
    
r_squared_value=coefficient(regression_line,y)
print r_squared_value
```
